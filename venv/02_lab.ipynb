```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\kerem\Desktop\pandas\data\nba.csv')

df.head()

```

```python
(
    df.groupby(by='Team')
    .agg({
        'Salary': 'sum'
    })
    .sort_values(
        by='Salary', 
        ascending=False
    )
    .head(10)
)
```

```python
(
    df.groupby(by='Team')
    .agg({
        'Age': 'mean'
    })
    .sort_values(
        by='Age', 
        ascending=False
    )
    .head(5)
)
df
```

```python
df_groupby_team = df.groupby('Team').mean()
df_groupby_team['Age'] = df_groupby_team['Age'].apply(lambda x: f'{x:.2f}')
df_groupby_team
```

```python
def str_find(name, prefix: str = 'and') -> bool:
  
    if isinstance(name, str):
        if prefix.lower() in name.lower():
            return True
    

    return False

df[
    df['Name'].apply(str_find)
]
```
