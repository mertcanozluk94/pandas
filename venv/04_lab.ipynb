```python
import pandas as pd
```

```python
customers = {
    'customer_id': [1,2,3,4,5],
    'user_name': [81,22,33,44,55],
    'address' : ['ank','ist','mrk','izm','burd']
}
orders = {
    'order_id': [4,5,6,7,8],
    'customer_id': [1,2,3,4,5],
    'order_revenue' : [2345,6789,1453,1299,4444]
}

df_customers=pd.DataFrame(customers)
df_orders=pd.DataFrame(orders)
```

```python
#inner merge
inner_merge = pd.merge(
        left=df_customers,
        right=df_orders,
        on='customer_id',
        how='inner')

inner_merge
```
