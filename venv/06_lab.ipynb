```python

df_can.replace('..', 0, inplace=True)


years = list(map(str, range(1980, 2014)))


df_can['Total'] = df_can[years].sum(axis=1)


by_continent = df_can.groupby('AreaName')['Total'].sum()


by_continent = by_continent[~by_continent.index.str.contains('Total')]


by_continent.plot(
    kind='pie',
    figsize=(12, 8),      
    startangle=90,
    autopct='%1.1f%%',
    shadow=True,
    labels=None,          
    pctdistance=1.12,     
    explode=[0.05] * len(by_continent)
)

plt.title(
    label='Immigration to Canada by Continent (1980–2013)',
    fontdict={'fontfamily': 'Times New Roman', 'fontsize': 18}
)
plt.ylabel('') 
plt.axis('equal') 

plt.legend(labels=by_continent.index, loc='upper left', bbox_to_anchor=(0.95, 1.0))

plt.show()
```
