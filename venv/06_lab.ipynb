```python
years: list[str] = list(map(str,range(1980,2014)))
df_changed_colrows = df_top_5_country[years].transpose()
print(df_changed_colrows.head())
```

```python
df_changed_colrows.style.set_caption("titlecan") #pip install jinja2

df_changed_colrows.plot(
    kind='area',
    figsize=(15,10),
    stacked=False,
    alpha=0.25,
)
plt.title(
    label='titlecan',
    fontdict={
        'fontfamily': 'TimesNewRoman',
        'fontsize' : '20'
    },
    color='red'
)
plt.xlabel(
    xlabel='Years',
    fontdict={
        'fontfamily': 'TimesNewRoman',
        'fontsize' : '20'
    },
    color='green'
)
plt.ylabel(
    ylabel='Area',
    fontdict={
        'fontfamily': 'TimesNewRoman',
        'fontsize' : '20'
    },
    color='green'
)
plt.show()
```
