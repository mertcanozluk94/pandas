```python
import pandas as pd
import numpy as np

df = pd.read_csv(r'C:\Users\kerem\Desktop\pandas\data\imdb.csv')

df.head()
```

```python
(
    df.query('Year > 2000')
            .loc[:,['Year','Votes']]
            .sort_values(by='Year',ascending = True)
)
```

```python
(
    df[df['Year'] > 2000]
    .sort_values(by='Rating', ascending=False)
    [['Title', 'Year', 'Director']]
    .head(12)
)
```

```python
(
    df.query('Year > 2000')
        .sort_values(by='Rating', ascending=False)
        [['Title', 'Year', 'Director']]
    .head(12)
)
```

```python
(
    df.query('2000 < Year < 2025')
        .loc[: , ['Title','Rating','Director']]
        .sort_values(
        by='Director',
        ascending=True
    
    
)
)
```

```python
(
    df[
        (df['Title'].str.contains('King', na=False)) & 
        (df['Rating'] > 7.0)
    ]
    [['Title', 'Rating', 'Director']]
    .sort_values(by='Rating', ascending=False)
)
```

```python
df[
    (df['Title'].str.contains('King')) & (df['Rating'] > 7.0)
][['Title', 'Rating', 'Actors']].sort_values(by='Rating', ascending=False)
```

```python
#path1 
df.groupby(by='Year')['Votes'].sum

#path2
(
    df.groupby(by='Year')
        .agg({
            'Votes':sum
        })
        .sort_values(
            by='Votes',
            ascending=False
        )
)
```
